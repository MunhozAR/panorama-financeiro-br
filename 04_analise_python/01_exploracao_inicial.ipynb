```python
import os
import glob

# Lista dos 10 bancos com nomes EXATOS do IF.data
mapa_bancos = {
    'ITAU': 'Itaú',
    'BRADESCO': 'Bradesco',
    'BB': 'Banco do Brasil',
    'SANTANDER': 'Santander',
    'NUBANK': 'Nubank',
    'C6 BANK': 'C6 Bank',
    'INTERMEDIUM': 'Inter',
    'PAGBANK-PAGSEGURO': 'PagBank',
    'MERCADO PAGO IP - FINANCEIRO': 'Mercado Pago',
    'PICPAY - FINANCEIRO': 'PicPay'
}

# Categoria de cada banco
categoria_bancos = {
    'Itaú': 'Tradicional',
    'Bradesco': 'Tradicional',
    'Banco do Brasil': 'Tradicional',
    'Santander': 'Tradicional',
    'Nubank': 'Digital',
    'C6 Bank': 'Digital',
    'Inter': 'Digital',
    'PagBank': 'Fintech Pagamento',
    'Mercado Pago': 'Fintech Pagamento',
    'PicPay': 'Fintech Pagamento'
}

# Lista todos os arquivos CSV
arquivos = sorted(glob.glob('../01_dados_brutos/ifdata_*.csv'))
print(f"Encontrei {len(arquivos)} arquivos:")
for arq in arquivos:
    print(f"  - {os.path.basename(arq)}")
```

```python
def ler_csv_com_trimestre(caminho):
    """
    Lê um CSV do IF.data e adiciona a coluna 'Trimestre'
    extraída do nome do arquivo (ex: 2025T4).
    """
    nome_arquivo = os.path.basename(caminho)
    trimestre = nome_arquivo.replace('ifdata_', '').replace('.csv', '')
    
    df = pd.read_csv(caminho, sep=';', encoding='utf-8-sig')
    df['Trimestre'] = trimestre
    
    return df

# Teste com 1 arquivo
df_teste = ler_csv_com_trimestre('../01_dados_brutos/ifdata_2025T4.csv')
print(f"Linhas: {len(df_teste)}")
print(f"Trimestre da primeira linha: {df_teste['Trimestre'].iloc[0]}")
print(f"Última coluna adicionada: {df_teste.columns[-1]}")
```

```python
# Lê os 8 arquivos e empilha tudo em um único DataFrame
lista_dfs = [ler_csv_com_trimestre(arq) for arq in arquivos]
df_completo = pd.concat(lista_dfs, ignore_index=True)

print(f"📊 Base unificada criada!")
print(f"   Total de linhas: {len(df_completo):,}")
print(f"   Total de colunas: {len(df_completo.columns)}")
print(f"   Trimestres: {sorted(df_completo['Trimestre'].unique())}")
```

```python
# Filtra só as instituições que estão no nosso mapa
df_bancos = df_completo[df_completo['Instituição'].isin(mapa_bancos.keys())].copy()

# Padroniza nome do banco
df_bancos['Banco'] = df_bancos['Instituição'].map(mapa_bancos)

# Adiciona categoria
df_bancos['Categoria'] = df_bancos['Banco'].map(categoria_bancos)

# Seleciona as colunas que importam
colunas_finais = [
    'Banco', 'Categoria', 'Trimestre', 'Data',
    'Ativo Total', 'Carteira de Crédito', 'Captações',
    'Patrimônio Líquido', 'Lucro Líquido', 'Número de Agências'
]
df_bancos = df_bancos[colunas_finais]

# Ordena e reseta índice
df_bancos = df_bancos.sort_values(['Banco', 'Trimestre']).reset_index(drop=True)

print(f"✅ Base filtrada!")
print(f"   Linhas: {len(df_bancos)}")
print(f"   Bancos: {sorted(df_bancos['Banco'].unique())}")
print(f"\nTrimestres por banco:")
print(df_bancos.groupby('Banco')['Trimestre'].count().sort_values(ascending=False))
```

```python
# Mapa ATUALIZADO com nomes alternativos
# Decisão analítica: empilhar razões sociais que representam a mesma instituição
mapa_bancos = {
    'ITAU': 'Itaú',
    'BRADESCO': 'Bradesco',
    'BB': 'Banco do Brasil',
    'SANTANDER': 'Santander',
    'NUBANK': 'Nubank',
    'C6 BANK': 'C6 Bank',
    'INTERMEDIUM': 'Inter',                                    # razão social antiga
    'INTER': 'Inter',                                          # razão social atual
    'PAGBANK-PAGSEGURO': 'PagBank',
    'MERCADO PAGO INSTITUIÇÃO DE PAGAMENTO LTDA.': 'Mercado Pago',  # entidade pagamento (antigo)
    'MERCADO PAGO IP - FINANCEIRO': 'Mercado Pago',                 # entidade financeira (recente)
    'PICPAY - FINANCEIRO': 'PicPay'
}

# Re-filtra com o mapa atualizado
df_bancos = df_completo[df_completo['Instituição'].isin(mapa_bancos.keys())].copy()
df_bancos['Banco'] = df_bancos['Instituição'].map(mapa_bancos)
df_bancos['Categoria'] = df_bancos['Banco'].map(categoria_bancos)

colunas_finais = [
    'Banco', 'Categoria', 'Trimestre', 'Data',
    'Ativo Total', 'Carteira de Crédito', 'Captações',
    'Patrimônio Líquido', 'Lucro Líquido', 'Número de Agências'
]
df_bancos = df_bancos[colunas_finais]
df_bancos = df_bancos.sort_values(['Banco', 'Trimestre']).reset_index(drop=True)

print(f"✅ Base re-filtrada!")
print(f"   Linhas: {len(df_bancos)}")
print(f"\nTrimestres por banco:")
print(df_bancos.groupby('Banco')['Trimestre'].count().sort_values(ascending=False))
```

```python
# Procura o trimestre duplicado do Mercado Pago
mp = df_bancos[df_bancos['Banco'] == 'Mercado Pago']
print("Mercado Pago por trimestre:")
print(mp[['Trimestre', 'Ativo Total', 'Patrimônio Líquido']].to_string(index=False))
print(f"\nQuantos registros por trimestre:")
print(mp.groupby('Trimestre').size())
```

```python
# Mapa FINAL — Opção C: separar Mercado Pago em duas entidades distintas
mapa_bancos = {
    'ITAU': 'Itaú',
    'BRADESCO': 'Bradesco',
    'BB': 'Banco do Brasil',
    'SANTANDER': 'Santander',
    'NUBANK': 'Nubank',
    'C6 BANK': 'C6 Bank',
    'INTERMEDIUM': 'Inter',
    'INTER': 'Inter',
    'PAGBANK-PAGSEGURO': 'PagBank',
    'MERCADO PAGO INSTITUIÇÃO DE PAGAMENTO LTDA.': 'Mercado Pago (Pagamento)',
    'MERCADO PAGO IP - FINANCEIRO': 'Mercado Pago (Financeiro)',
    'PICPAY - FINANCEIRO': 'PicPay'
}

# Categoria atualizada para as duas entidades do Mercado Pago
categoria_bancos = {
    'Itaú': 'Tradicional',
    'Bradesco': 'Tradicional',
    'Banco do Brasil': 'Tradicional',
    'Santander': 'Tradicional',
    'Nubank': 'Digital',
    'C6 Bank': 'Digital',
    'Inter': 'Digital',
    'PagBank': 'Fintech Pagamento',
    'Mercado Pago (Pagamento)': 'Fintech Pagamento',
    'Mercado Pago (Financeiro)': 'Fintech Pagamento',
    'PicPay': 'Fintech Pagamento'
}

# Re-filtra
df_bancos = df_completo[df_completo['Instituição'].isin(mapa_bancos.keys())].copy()
df_bancos['Banco'] = df_bancos['Instituição'].map(mapa_bancos)
df_bancos['Categoria'] = df_bancos['Banco'].map(categoria_bancos)

colunas_finais = [
    'Banco', 'Categoria', 'Trimestre', 'Data',
    'Ativo Total', 'Carteira de Crédito', 'Captações',
    'Patrimônio Líquido', 'Lucro Líquido', 'Número de Agências'
]
df_bancos = df_bancos[colunas_finais]
df_bancos = df_bancos.sort_values(['Banco', 'Trimestre']).reset_index(drop=True)

print(f"✅ Base FINAL!")
print(f"   Linhas: {len(df_bancos)}")
print(f"\nTrimestres por banco:")
print(df_bancos.groupby('Banco')['Trimestre'].count().sort_values(ascending=False))
```

```python
# Lista das colunas numéricas que precisam ser convertidas
colunas_numericas = [
    'Ativo Total', 'Carteira de Crédito', 'Captações',
    'Patrimônio Líquido', 'Lucro Líquido', 'Número de Agências'
]

# Para cada coluna: remove os pontos (separador de milhar) e converte para float
for col in colunas_numericas:
    df_bancos[col] = (
        df_bancos[col]
        .astype(str)              # garante que está como string
        .str.replace('.', '', regex=False)  # remove pontos de milhar
        .str.replace(',', '.', regex=False) # vírgula decimal → ponto decimal
        .replace('', '0')         # casos vazios viram zero
        .astype(float)            # converte para número
    )

# Verifica os tipos depois da conversão
print("Tipos das colunas após conversão:")
print(df_bancos[colunas_numericas].dtypes)

print("\nExemplo de valor numérico em 'Lucro Líquido':")
print(df_bancos['Lucro Líquido'].iloc[0])

print("\nPrimeiras linhas (só para conferir):")
print(df_bancos[['Banco', 'Trimestre', 'Ativo Total', 'Lucro Líquido']].head(3).to_string(index=False))
```

```python
# Calcula ROE anualizado (Lucro Líquido / Patrimônio Líquido) × 4
df_bancos['ROE_anualizado'] = (df_bancos['Lucro Líquido'] / df_bancos['Patrimônio Líquido']) * 4
df_bancos['ROE_anualizado'] = df_bancos['ROE_anualizado'].round(4)

# Salva a base consolidada
caminho_saida = '../02_dados_tratados/base_consolidada.csv'
df_bancos.to_csv(caminho_saida, sep=';', encoding='utf-8-sig', index=False)

print(f"✅ Base consolidada salva em: {caminho_saida}")
print(f"   Linhas: {len(df_bancos)}")
print(f"   Colunas: {list(df_bancos.columns)}")

# Prévia: ROE médio por banco no último trimestre
print(f"\n📊 ROE no trimestre mais recente (2025T4):")
ultimo = df_bancos[df_bancos['Trimestre'] == '2025T4'][['Banco', 'Categoria', 'ROE_anualizado']]
ultimo = ultimo.sort_values('ROE_anualizado', ascending=False)
print(ultimo.to_string(index=False))
```
